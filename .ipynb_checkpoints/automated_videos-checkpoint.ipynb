```python
def video_stats(api_keys, video_ids, endpoint):
    '''
    Function to retrieve the video statistics for certain video_ids via the API
    '''
    # Initialize an empty dictionary to store the results
    vid_stats = {}
    total_keys = len(api_keys)
    current_key_index = 0

    # Function to make a request with the current API key
    def make_request(api_key, video_id):
        params = {
            'part': 'snippet,statistics,contentDetails',
            'key': api_key,
            'id': video_id
        }
        return requests.get(endpoint, params=params)

    # Iterate over video IDs with a progress bar
    for video_id in tqdm(video_ids, desc="Processing videos"):
        success = False
        
        while not success:
            if current_key_index >= total_keys:
                print("All keys exceeded quota. Sleeping for 1 day...")
                time.sleep(86400)  # Sleep for 1 day
                current_key_index = 0
            
            response = make_request(api_keys[current_key_index], video_id)
            
            if response.status_code == 200:
                data = response.json()
                if 'error' not in data:
                    vid_stats[video_id] = data
                    success = True
                elif 'quotaExceeded' in data['error'].get('message', ''):
                    print(f"Quota exceeded for key {api_keys[current_key_index]}. Trying next key...")
                    current_key_index += 1
            else:
                print(f"Error retrieving info for video ID: {video_id}")
                break

    return vid_stats

def update_csv_with_video_stats(df, stats):
    video_published_all = []
    descriptions_all = []
    tags_all = []
    categories_all = []
    live_all = []
    durations = []
    captions = []
    views = []
    likes = []
    favs = []
    comments = []

    # Extract data from stats
    for key, values in stats.items():
        try:
            item = values["items"][0]
            snippet = item["snippet"]
            content_details = item["contentDetails"]
            statistics = item["statistics"]
            
            video_published_all.append(snippet['publishedAt'])
            descriptions_all.append(snippet['description'])
            tags_all.append(snippet.get('tags', []))  # Use .get to handle missing tags
            categories_all.append(snippet['categoryId'])
            live_all.append(snippet['liveBroadcastContent'])
            durations.append(content_details['duration'])
            captions.append(content_details['caption'])
            views.append(statistics['viewCount'])
            likes.append(statistics['likeCount'])
            favs.append(statistics['favoriteCount'])
            try:
                comments.append(statistics['commentCount'])
            except:
                comments.append(None)
        except:
            video_published_all.append(None)
            descriptions_all.append(None)
            tags_all.append(None)
            categories_all.append(None)
            live_all.append(None)
            durations.append(None)
            captions.append(None)
            views.append(None)
            likes.append(None)
            favs.append(None)
            comments.append(None)     

    # Check if the lengths match
    if len(video_published_all) == len(df):
        # Add new data to the dataframe
        df['video_published'] = video_published_all
        df['descriptions'] = descriptions_all
        df['tags'] = tags_all
        df['categories'] = categories_all
        df['live'] = live_all
        df['durations'] = durations
        df['captions'] = captions
        df['views'] = views
        df['likes'] = likes
        df['favorites'] = favs
        df['comments'] = comments

        # Save the updated dataframe back to CSV
        return df
    else:
        print(len(video_published_all))
        print("Error: The number of video stats does not match the number of rows in the CSV.")
        
        
# Function to convert ISO 8601 duration to minutes
def iso_to_minutes(iso_duration):
    duration = isodate.parse_duration(iso_duration)
    return duration.total_seconds() / 60
```

```python
df_videos = pd.read_csv("videos.csv")
```

```python
# Retrieve video statistics
video_ids = df_videos['video_id'].tolist()

stats = video_stats(keys, video_ids, video_stats_endpoint)

# Update the DataFrame with video statistics
df_videos_updated = update_csv_with_video_stats(df_videos, stats)

# Convert ISO 8601 duration to minutes
df_videos_updated['durations'] = df_videos_updated['durations'].fillna('PT0M0S')
df_videos_updated['duration_minutes'] = df_videos_updated['durations'].apply(iso_to_minutes)

# Save the updated DataFrame to CSV
df_videos_updated.to_csv("videos_all.csv", index=False)

print("Data collection complete and saved to videos_all.csv")
```
