```python
import requests
import time
import pandas as pd
from tqdm import tqdm
```

```python
# Set the API endpoint
comment_endpoint = "https://www.googleapis.com/youtube/v3/commentThreads"

# Set the API endpoint
replies_endpoint = "https://www.googleapis.com/youtube/v3/comments"
```

```python
# define private api_keys 


keys = [
    key1,
    key2,
    key3,
    key4,
    key5,
    key6,
    key7,
    key8,
    key9,
    key10,
    key11,
    key12,
    key13,
    key14,
    key15,
    key16,
    key17,
    key18,
    key19,
    key20
]
```

```python
def comment_request(api_keys, video_ids, endpoint, replies_endpoint):
    '''
    Function to retrieve the comments for certain video_ids via the API
    '''
    comments_dict = {}
    total_keys = len(api_keys)
    current_key_index = 0

    # Function to make a request with the current API key
    def make_request(api_key, params):
        return requests.get(endpoint, params=params)

    # Iterate over the video IDs with a progress bar
    for video_id in tqdm(video_ids, desc="Processing videos for comments"):
        success = False
        while not success:
            if current_key_index >= total_keys:
                print("All keys exceeded quota. Sleeping for 1 day...")
                time.sleep(86400)  # Sleep for 1 day
                current_key_index = 0

            params = {
                "key": api_keys[current_key_index],
                "videoId": video_id,
                "part": "id,snippet,replies",
                "maxResults": 100  # Set the desired number of comments per video
            }

            response = make_request(api_keys[current_key_index], params)

            if response.status_code == 200:
                data = response.json()
                comment_thread = data["items"]
                while "nextPageToken" in data:
                    next_page_token = data["nextPageToken"]
                    params["pageToken"] = next_page_token
                    response = make_request(api_keys[current_key_index], params)
                    if response.status_code == 200:
                        data = response.json()
                        comment_thread += data["items"]
                    elif response.status_code == 403:
                        print(f"Quota exceeded for key {api_keys[current_key_index]} during pagination. Trying next key...")
                        current_key_index += 1
                        if current_key_index >= total_keys:
                            print("All keys exceeded quota during pagination. Sleeping for 1 day...")
                            time.sleep(86400)  # Sleep for 1 day
                            current_key_index = 0
                        params["key"] = api_keys[current_key_index]
                    else:
                        print(f"Error occurred while fetching comments during pagination: {response.status_code}")
                        break
                success = True
            elif response.status_code == 403:
                print(f"Quota exceeded for key {api_keys[current_key_index]}. Trying next key...")
                current_key_index += 1
            else:
                print(f"Error occurred while fetching comments for video ID {video_id}: {response.status_code}")
                break

        for i in comment_thread:
            if i['snippet']['totalReplyCount'] > 0:
                if i['snippet']['totalReplyCount'] > len(i['replies']["comments"]):           
                    replies_params = {
                        "key": api_keys[current_key_index],
                        "part": "snippet",
                        "parentId": i["id"]
                    }
                    replies_response = requests.get(replies_endpoint, params=replies_params)
                    data = replies_response.json()
                    i["replies"]["comments"] = data

        comments_dict[video_id] = comment_thread

    return comments_dict

def process_comments(comments_dict, channel_ids):
    text = []
    publishedAt = []
    authorID = []
    likeCount = []
    totalReplyCount = []
    parentID = []
    commentID = []
    videoID = []
    channelId = []
    username = []

    for video_id, comments in comments_dict.items():
        channel_id = channel_ids.get(video_id)
        for i in comments:
            comment_id = i["id"]
            try:
                comment_data = {
                    "publishedAt": i["snippet"]['topLevelComment']["snippet"]['publishedAt'],
                    "text": i["snippet"]['topLevelComment']["snippet"]['textOriginal'],
                    "authorID": i["snippet"]['topLevelComment']["snippet"]['authorChannelId']['value'],
                    "likeCount": i["snippet"]['topLevelComment']["snippet"]['likeCount'],
                    "totalReplyCount": i["snippet"]['totalReplyCount'],
                    "parentID": 0
                }

                text.append(i["snippet"]['topLevelComment']["snippet"]['textOriginal'])
                publishedAt.append(i["snippet"]['topLevelComment']["snippet"]['publishedAt'])
                authorID.append(i["snippet"]['topLevelComment']["snippet"]['authorChannelId']['value'])
                likeCount.append(i["snippet"]['topLevelComment']["snippet"]['likeCount'])
                totalReplyCount.append(i["snippet"]['totalReplyCount'])
                parentID.append(0)
                commentID.append(i["id"])
                videoID.append(video_id)
                channelId.append(channel_id)
                username.append(i["snippet"]['topLevelComment']["snippet"]['authorDisplayName'])
            except:
                comment_data = {
                    "publishedAt": i["snippet"]['topLevelComment']["snippet"]['publishedAt'],
                    "text": i["snippet"]['topLevelComment']["snippet"]['textOriginal'],
                    "authorID": 0,
                    "likeCount": i["snippet"]['topLevelComment']["snippet"]['likeCount'],
                    "totalReplyCount": i["snippet"]['totalReplyCount'],
                    "parentID": 0
                }

                text.append(i["snippet"]['topLevelComment']["snippet"]['textOriginal'])
                publishedAt.append(i["snippet"]['topLevelComment']["snippet"]['publishedAt'])
                authorID.append(0)
                likeCount.append(i["snippet"]['topLevelComment']["snippet"]['likeCount'])
                totalReplyCount.append(i["snippet"]['totalReplyCount'])
                parentID.append(0)
                commentID.append(i["id"])
                videoID.append(video_id)
                channelId.append(channel_id)
                username.append(i["snippet"]['topLevelComment']["snippet"]['authorDisplayName'])

            # Check for replies and extract them
            if i["snippet"]['totalReplyCount'] > 0:
                for x in i["replies"]["comments"]:
                    try:
                        reply_id = x["id"]
                        reply_data = {
                            "publishedAt": x["snippet"]['publishedAt'],
                            "text": x["snippet"]['textOriginal'],
                            "authorID": x["snippet"]['authorChannelId']['value'],
                            "likeCount": x["snippet"]['likeCount'],
                            "totalReplyCount": 0,
                            "parentID": x["snippet"]['parentId']
                        }

                        text.append(x["snippet"]['textOriginal'])
                        publishedAt.append(x["snippet"]['publishedAt'])
                        authorID.append(x["snippet"]['authorChannelId']['value'])
                        likeCount.append(x["snippet"]['likeCount'])
                        totalReplyCount.append(0)
                        parentID.append(x["snippet"]['parentId'])
                        commentID.append(reply_id)
                        videoID.append(video_id)
                        channelId.append(channel_id)
                        username.append(x["snippet"]['authorDisplayName'])
                    except:
                        pass

    df_comments = pd.DataFrame({
        'video_id': videoID,
        'comment_id': commentID,
        'comment_text': text,
        'comment_author': username,
        'comment_published_at': publishedAt,
        'reply_to': parentID,
        'author_id': authorID,
        'like_count': likeCount,
        'total_reply_count': totalReplyCount,
        'channel_id': channelId
    })

    return df_comments
```

```python
df_videos = pd.read_csv("videos_all.csv")
```

```python
# Assume df_videos is your DataFrame containing video IDs and channel IDs
video_ids = df_videos['video_id'].tolist()
channel_ids = dict(zip(df_videos['video_id'], df_videos['channel_id']))

# Retrieve comments
comments_dict = comment_request(keys, video_ids, comments_endpoint, replies_endpoint)

# Process comments into a DataFrame
df_comments = process_comments(comments_dict, channel_ids)

# Save the comments DataFrame to CSV
df_comments.to_csv("comments_all.csv", index=False)

print("Comments data collection complete and saved to comments_all.csv")
```
